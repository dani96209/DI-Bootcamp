```python
df.rename(columns={'iyear':'Year','imonth':'Month','iday':'Day','country_txt':'Country',
                       'region_txt':'Region','attacktype1_txt':'AttackType','target1':'Target',
                       'nkill':'Killed','nwound':'Wounded','summary':'Summary','gname':'Group',
                       'targtype1_txt':'Target_type','weaptype1_txt':'Weapon_type','motive':'Motive'},inplace=True)
```

Some variables have been modified for better readability

```python
df.head()
```

```python
df=df[['Year','Month','Day','Country','Region','city','latitude','longitude','AttackType','Killed',
       'Wounded','Target','extended','Group','Target_type','Weapon_type','Motive']]
```

As said before, there are many variables, some of which are useless, we decide to keep some variables which will be very useful in our analysis.

```python
df.info()
```

```python
missing_percentage = df.isnull().sum() / len(df) * 100
print(missing_percentage)
```

Most of the values ​​of the 'Motive' variable are unavailable, so we decide to remove this variable from our dataset.

And we fill the values ​​of the other variables with the mode and the median value.

```python
df['city'].fillna(df['city'].mode()[0],inplace=True)
df['AttackType'].fillna(df['AttackType'].mode()[0],inplace=True)
df['Target'].fillna(df['Target'].mode()[0],inplace=True)
df.drop(columns=['Motive'],inplace=True)
```

```python
df['Killed'].fillna(df['Killed'].median(),inplace=True)
df['Wounded'].fillna(df['Wounded'].median(),inplace=True)
```

```python
df['city'] = df['city'].str.strip().str.lower()  # Supprime les espaces et uniformise
```

```python
df['latitude'] = df.groupby('Country')['latitude'].transform(lambda x: x.fillna(x.median()))
df['longitude'] = df.groupby('Country')['longitude'].transform(lambda x: x.fillna(x.median()))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
# Correlation matrix with all numerical variables
numeric_vars = ['Year', 'Month', 'Day', 'latitude', 'longitude', 'Killed', 'Wounded', 'extended']
df_numeric = df[numeric_vars]


corr_matrix = df_numeric.corr()

# Display the matrix as a heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)

# Add a title
plt.title("Correlation matrix of numerical variables")
plt.show()
```

The only most correlated variables are wounded with killed and latitude with year.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Total number of people killed and injured
total_killed = df['Killed'].sum()
total_wounded = df['Wounded'].sum()

# Creating the DataFrame for Visualization
data = {'Category': ['Killed', 'Wounded'], 'Number': [total_killed, total_wounded]}
df_victims = pd.DataFrame(data)

# Creating the chart
plt.figure(figsize=(8, 6))
sns.barplot(x='Category', y='Number', data=df_victims, palette=['red', 'blue'])

# Add the values ​​above the bars
for i, value in enumerate(df_victims['Number']):
    plt.text(i, value + 100, str(int(value)), ha='center', fontsize=12, fontweight='bold')

# Add a title and labels
plt.title("Total number of people killed and injured in the attacks", fontsize=14)
plt.xlabel("")
plt.ylabel("Victims Number")

# Show chart
plt.show()
```

More people are injured in attacks than killed.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Aggregate data by country
df_country_victims = df.groupby('Country')[['Killed', 'Wounded']].sum().reset_index()

# Sort by total number of casualties (killed + injured) and take the first 20
df_country_victims['Total Victims'] = df_country_victims['Killed'] + df_country_victims['Wounded']
df_top_20_countries = df_country_victims.sort_values(by="Total Victims", ascending=False).head(20)

# Create a DataFrame for better visualization
df_top_20_melted = df_top_20_countries.melt(id_vars="Country", value_vars=["Killed", "Wounded"],
                                             var_name="Category", value_name="Number")

# Creating the chart
plt.figure(figsize=(14, 8))
sns.barplot(x="Number", y="Country", hue="Category", data=df_top_20_melted, palette=["red", "blue"])

# Add a title and labels
plt.title("Total number of people killed and injured in the 20 most affected countries", fontsize=14)
plt.xlabel("Victims Number")
plt.ylabel("Country")

# Afficher le graphique
plt.show()
```

It can be seen that the number of injured is significantly higher than the number of deaths for countries such as Iraq, Pakistan and India. On the other hand, for some countries recording fewer attacks, the order is reversed with a number of people killed higher than the number of people injured.

To gain a better understanding of terrorism operating worldwide, we transformed our initial dataset to retain only the most essential variables and remove variables that complicated the analysis.

We coded certain categorical variables to be able to perform a correlation matrix analysis with the numerical variables. However, we did not code all the categorical variables because it would have become illegible during the correlation matrix visualization due to the very large number of modalities for certain categorical variables.

We filled the missing values ​​for some variables using the mode and for others using the median value. Based on all of these transformations, we were able to conduct our various analyses more easily, with different results observed on terrorism:

First of all, the countries and regions most affected by terrorism are the Middle East, North Africa, and South Asia.

Among the victims of these terrorist attacks, more people are injured than killed.

The weapons used in terrorist operations are explosives, firearms, and fire.
