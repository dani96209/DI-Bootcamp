```python
# Exercise 1:
import pandas as pd

titanic_data=pd.read_csv('train.csv')
duplicate_rows = titanic_data[titanic_data.duplicated()]
print(titanic_data.duplicated().sum())
print(f"Nombre de lignes dupliquées : {len(duplicate_rows)}")
df_cleaned = titanic_data.drop_duplicates()
print(f"Nombre de lignes avant suppression : {len(titanic_data)}")
print(f"Nombre de lignes après suppression : {len(df_cleaned)}")
titanic_data.head()
titanic_data.shape
titanic_data = titanic_data.drop_duplicates()

titanic_data.shape
```

```python
#Exercise 2:
# Check for missing values in a DataFrame
from sklearn.impute import SimpleImputer

missing_data = titanic_data.isnull()
print(missing_data.head())

# Count missing values in each column

titanic_data = titanic_data.dropna(subset=['Embarked'])

titanic_data["Cabin"].fillna("Unknown",inplace=True)

imputer = SimpleImputer(strategy="mean")
titanic_data["Age"] = imputer.fit_transform(titanic_data[["Age"]])
print(titanic_data.isnull().sum())  # Vérifier que les valeurs manquantes ont été traitées
```

```python
#Exercise3:
from sklearn.preprocessing import LabelEncoder

titanic_data["Family_Size"] = titanic_data["SibSp"] + titanic_data["Parch"] + 1

# Extraction des titres depuis la colonne "Name"
titanic_data["Title"] = titanic_data["Name"].str.extract(r' ([A-Za-z]+)\.', expand=False)

# Affichage des premières lignes pour vérification
print(titanic_data[["Name", "Title", "Family_Size"]].head())
titanic_data = pd.get_dummies(titanic_data, columns=["Title"], drop_first=True)

# Label encoding (alternative, applique un numéro unique à chaque catégorie)
le = LabelEncoder()
titanic_data["Sex"] = le.fit_transform(titanic_data["Sex"])
from sklearn.preprocessing import MinMaxScaler, StandardScaler

scaler = MinMaxScaler()
titanic_data[["Age", "Fare", "Family_Size"]] = scaler.fit_transform(titanic_data[["Age", "Fare", "Family_Size"]])
print(titanic_data.head())

```

```python
# Exercise 4:
def detect_outliers_iqr(titanic_data, column):
    Q1 = titanic_data[column].quantile(0.25)
    Q3 = titanic_data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return lower_bound, upper_bound

# Outliers détectés pour Age et Fare
lower_bound_age, upper_bound_age = detect_outliers_iqr(titanic_data, "Age")
lower_bound_fare, upper_bound_fare = detect_outliers_iqr(titanic_data, "Fare")

# Filtrer les données en supprimant les outliers pour Age et Fare
titanic_data_cleaned = titanic_data[
    (titanic_data['Age'] >= lower_bound_age) & (titanic_data['Age'] <= upper_bound_age) &
    (titanic_data['Fare'] >= lower_bound_fare) & (titanic_data['Fare'] <= upper_bound_fare)
]

# Calculer le nombre d'outliers pour chaque colonne (Age et Fare)
outliers_age = titanic_data[(titanic_data['Age'] < lower_bound_age) | (titanic_data['Age'] > upper_bound_age)]
outliers_fare = titanic_data[(titanic_data['Fare'] < lower_bound_fare) | (titanic_data['Fare'] > upper_bound_fare)]
# Afficher les résultats
print(f"Nombre d'outliers dans Age: {len(outliers_age)}")
print(f"Nombre d'outliers dans Fare: {len(outliers_fare)}")

size_data_before=titanic_data.shape
print(size_data_before)
size_data_after=titanic_data_cleaned.shape
print(size_data_after)
```

```python
# Exercise 5:
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns

num_cols = ["Age", "Fare"]

# Afficher les statistiques descriptives
print(titanic_data[num_cols].describe())

scaler = StandardScaler()
titanic_data_standardized = titanic_data.copy()

# Appliquer StandardScaler sur les colonnes numériques
titanic_data_standardized[num_cols] = scaler.fit_transform(titanic_data[num_cols])

# Vérifier les nouvelles statistiques
print(titanic_data_standardized[num_cols].describe())

plt.figure(figsize=(12, 5))
for i, col in enumerate(num_cols):
    plt.subplot(1, len(num_cols), i + 1)
    sns.histplot(titanic_data_standardized[col], kde=True)
    plt.title(f"Distribution Standardisée de {col}")
plt.show()

#Normalisation:
scaler_minmax = MinMaxScaler()
titanic_data_normalized = titanic_data.copy()

# Appliquer MinMaxScaler sur les colonnes numériques
titanic_data_normalized[num_cols] = scaler_minmax.fit_transform(titanic_data[num_cols])

# Vérifier les nouvelles statistiques
print(titanic_data_normalized[num_cols].describe())

# Visualisation après normalisation
plt.figure(figsize=(12, 5))
for i, col in enumerate(num_cols):
    plt.subplot(1, len(num_cols), i + 1)
    sns.histplot(titanic_data_normalized[col], kde=True)
    plt.title(f"Distribution Normalisée de {col}")
plt.show()
```

```python
# Dailychallenge:
import pandas as pd
from sklearn.preprocessing import LabelEncoder

from sklearn.preprocessing import MinMaxScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
data_scientist_salary=pd.read_csv('datascience_salaries.csv',sep=',')
print(data_scientist_salary.columns)

data_scientist_salary.head()
label_encoder = LabelEncoder()
data_scientist_salary['job_type'] = label_encoder.fit_transform(data_scientist_salary['job_type'])
data_scientist_salary['location'] = label_encoder.fit_transform(data_scientist_salary['location'])
data_scientist_salary.head()

scaler = MinMaxScaler()
data_scientist_salary['normalized_salary'] = scaler.fit_transform(data_scientist_salary[['salary']])
print(data_scientist_salary[['salary', 'normalized_salary']].head())
variables = ['job_type','location']
pca = PCA(n_components=2)  # Reduce to 2 dimensions for visualization
data_pca = pca.fit_transform(data_scientist_salary[variables])

# Add PCA results to the DataFrame
data_scientist_salary['PCA1'] = data_pca[:, 0]
data_scientist_salary['PCA2'] = data_pca[:, 1]

# 3. Group by 'experience_level' and calculate average and median salary
salary_stats = data_scientist_salary.groupby('experience_level')['salary'].agg(['mean', 'median'])

# Display the results
print("Normalized Salary and PCA Results:")
print(data_scientist_salary[['salary', 'normalized_salary', 'PCA1', 'PCA2']])
print("\nAverage and Median Salary by Experience Level:")
print(salary_stats)

# Plot PCA results for visualization
plt.scatter(data_scientist_salary['PCA1'], data_scientist_salary['PCA2'], c=data_scientist_salary['salary'], cmap='viridis')
plt.colorbar(label='Salary')
plt.xlabel('PCA1')
plt.ylabel('PCA2')
plt.title('PCA of Features')
plt.show()
```
