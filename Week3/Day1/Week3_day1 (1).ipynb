```python
# Exercise 3
#Quantitative (Numerical Data): Columns containing numeric values that can be measured or counted and used in mathematical operations. These can be:
#Discrete: Whole numbers (e.g., number of students, product count)
#Continuous: Measurable values (e.g., height, weight, temperature)
#Qualitative (Categorical Data): Columns containing non-numeric values representing categories, labels, or characteristics. These can be:
#Nominal: Categories with no inherent order (e.g., gender, country, eye color)
#Ordinal: Categories with a meaningful order but no uniform difference (e.g., satisfaction level, education level)
print(data_american_sleep.dtypes)
print(data_mentalhealth.dtypes)
print(data_creditcard.dtypes)
```

```python
# Exercise 5
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

print(data_iris.head())
mean_value = np.mean(data_iris["SepalLengthCm"])
median_value = np.median(data_iris["PetalLengthCm"])
print(mean_value)
print(median_value)
mode_value = stats.mode(data_iris["PetalLengthCm"]).mode
print(mode_value)

plt.plot(data_iris['SepalLengthCm'], data_iris['SepalWidthCm'])
plt.xlabel('SepalLengthCm')
plt.ylabel('SepalWidthCm')
plt.show()

# Observations:
#The observed average of the different iris species regarding the length of the sepals is about 5.84 cm.
# With the calculation of the median, it can be seen that 50% of the flower species have a petal length greater than 4.35 cm and 50% have a petal length less than 4.35 cm.

# The most frequent observed value regarding the length of the petals is 1.5 cm.

 # The graph describes a relationship between the length and the width of the sepals of all the iris spaces. However this graph clearly highlights a clear separation between the iris species with some species having a smaller width and length of sepals compared to other species.
```

In the COVID-19 Pandemic, data analysis played a critical role in shaping public health strategies, informing policy, and guiding healthcare decisions.

Data Analyzed:
Infection Rates: Data on infection rates, positive cases, and recovery rates were collected from hospitals, health departments, and global tracking platforms.
Hospital Capacity: Data from hospitals provided insights into available beds, ICU usage, and the demand for ventilators.
Mortality Rates: The global death toll and trends helped shape responses like lockdowns and quarantines.
Vaccination Data: Data on vaccine efficacy, distribution, and rollout played a role in prioritizing certain groups (e.g., elderly, healthcare workers).
Methods Used:
Descriptive Statistics: Used to summarize and visualize case counts, mortality rates, and recovery trends.
Predictive Modeling: Used to forecast future case numbers, hospital demand, and death tolls based on various public health interventions.
Geospatial Analysis: Allowed for tracking regional outbreaks, hotspots, and mapping the spread of the virus.
Statistical Tests: Used in vaccine trials to assess the safety and efficacy of vaccines.

Without data analysis, decision-making during the pandemic would have been chaotic and reactive rather than proactive. Governments and organizations would not have been able to track the virus's spread effectively or forecast the healthcare needs that arose. For example:

Without real-time data on infection rates, lockdowns may have been delayed or mismanaged.
Hospitals might have faced resource shortages without data on ICU capacity and equipment needs.
How Did Data Analysis Contribute to Solving a Problem or Making a Decision?

Vaccine Rollout: By analyzing risk factors, data helped prioritize vulnerable populations for vaccination, improving the efficiency of vaccine distribution.
