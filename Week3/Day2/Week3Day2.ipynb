```python
# Load an image using PIL
from PIL import Image
import matplotlib.pyplot as plt
image_path = 'flowers/flowers/19_010.png'
original_image = Image.open(image_path)

# Display the original image using matplotlib
plt.imshow(original_image)
```

```python
from scipy.ndimage import rotate
image_array = np.array(original_image)

def rotate_image_30_degrees(image_array):
    return rotate(image_array, angle=30, reshape=True)

# Appliquer la rotation
rotated_image = rotate_image_30_degrees(image_array)

# Afficher l'image pivotée
plt.imshow(rotated_image)
plt.title("Image Rotated 30°")
plt.axis("off")
plt.show()
```

```python
from PIL import Image, ImageOps
# Retourner l'image horizontalement
flipped_horizontally = ImageOps.mirror(original_image)

# Retourner l'image verticalement
flipped_vertically = ImageOps.flip(original_image)

# Afficher les images retournées
fig, axes = plt.subplots(1, 2, figsize=(10, 5))

axes[0].imshow(flipped_horizontally)
axes[0].set_title("Retourner Horizontalement")
axes[0].axis("off")

axes[1].imshow(flipped_vertically)
axes[1].set_title("Retourner Verticalement")
axes[1].axis("off")

plt.show()
```

```python
def zoom_image(image, scale=1.2):
    width, height = image.size
    new_width = int(width * scale)
    new_height = int(height * scale)

    # Redimensionner l'image
    return image.resize((new_width, new_height), Image.LANCZOS)

# Appliquer le zoom
zoomed_image = zoom_image(original_image, scale=1.2)

# Afficher l'image zoomée
plt.imshow(zoomed_image)
plt.title("Image Zoomée (x1.2)")
plt.axis("off")
plt.show()
```
