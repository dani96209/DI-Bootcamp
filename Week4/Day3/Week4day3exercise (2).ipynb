```python
# Exercise 2:
from scipy import stats
import numpy as np

# Sample dataset
data = [12, 15, 13, 12, 18, 20, 22, 21]

#  statistics computation
mean_value = stats.tmean(data)
median_value = np.median(data)
variance_value = stats.tvar(data, ddof=1)  # ddof=1 pour l'échantillon (variance corrigée)
std_dev_value = stats.tstd(data, ddof=1)   # ddof=1 pour l'écart-type corrigé

# Results display
print(f"Mean: {mean_value:.2f}")
print(f"Median: {median_value}")
print(f"Variance: {variance_value:.2f}")
print(f"Standard Deviation: {std_dev_value:.2f}")
```

```python
# Exercise3:
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
mu, sigma = 50, 10  # mean and standard deviation
 # Creating the normal distribution object

normal_dist = norm(mu, sigma)

# Calculation of the probability density for x=0.5

pdf_value = normal_dist.pdf(0.5)
print("PDF at x=0.5:", pdf_value)

# Generate values ​​for the X axis
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)
pdf = normal_dist.pdf(x)

# PLot the Normal Distribution
plt.figure(figsize=(8, 5))
plt.plot(x, pdf, label="Normal Distribution", color='blue')
plt.fill_between(x, pdf, alpha=0.2, color='blue')  # Effet d'ombre sous la courbe
plt.axvline(mu, color='red', linestyle='dashed', label=f"Mean ({mu})")
plt.title("Normal Distribution (μ=50, σ=10)")
plt.xlabel("Values")
plt.ylabel("Probability Density (PDF)")
plt.legend()
plt.grid()
plt.show()

```

```python
# Exercise 4:
import numpy as np
from scipy.stats import ttest_ind

# # Generating two random data sets
data1 = np.random.normal(50, 10, 100)  # Moyenne = 50, Écart-type = 10, Taille = 100
data2 = np.random.normal(60, 10, 100)  # Moyenne = 60, Écart-type = 10, Taille = 100

# Perform an independent samples t-test
t_stat, p_value = ttest_ind(data1, data2)

# Results display
print(f"T-statistic: {t_stat:.2f}")
print(f"P-value: {p_value:.4f}")
```

```python
# Exercise 5:
import numpy as np
from scipy.stats import linregress

# Data
house_sizes = np.array([50, 70, 80, 100, 120])
house_prices = np.array([150000, 200000, 210000, 250000, 280000])

# Slope : Represents the rate of change of the dependent variable (e.g., house price) with respect to the independent variable (e.g., house size). It tells us how much the price increases for each additional square meter.
# Intercept : The value of the dependent variable when the independent variable is zero. It represents the estimated price when the house size is 0 .

# Linear regression
slope, intercept, r_value, p_value, std_err = linregress(house_sizes, house_prices)

print(f" (slope): {slope:.2f}")
print(f"Intercept: {intercept:.2f}")
```

```python
size_to_predict = 90
predicted_price = slope * size_to_predict + intercept
print(f"Estimated price for a 90 m² house : {predicted_price:.2f}")
# If the slope is, for example, 1828.77, this means that each additional m² increases the price by 1828.77 monetary units.
# Thus, the larger the house, the more its price increases proportionally
```

```python
# Exercise 6:
import scipy.stats as stats

# Données des trois engrais
fertilizer_1 = [5, 6, 7, 6, 5]
fertilizer_2 = [7, 8, 7, 9, 8]
fertilizer_3 = [4, 5, 4, 3, 4]

# Test ANOVA
f_value, p_value = stats.f_oneway(fertilizer_1, fertilizer_2, fertilizer_3)

print(f"F-value: {f_value:.2f}")
print(f"P-value: {p_value:.4f}")

# F-value: Measures the variation between groups relative to the variation within groups.More precisely, the F-value in ANOVA helps us determine the ratio of the variance between the groups to the variance within the groups. A high F-value typically indicates that there are significant differences between the means of the groups
# P-value: Probability that the observed differences are due to chance.

 # If the P-value is greater than alpha=0.05, it means that we cannot reject the null hypothesis and so  the fertilizers haven't significantly different effects on plant growth

```

Result:
P-value<0.05, so we reject the null hypothesis and we argue that the fertilizers have significantly different effects on plant growth
