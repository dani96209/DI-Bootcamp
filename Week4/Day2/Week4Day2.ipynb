```python
# Exercise 6:
import numpy as np
import scipy.stats as stats

# Productivity scores of employees before the training program
productivity_before = np.random.normal(loc=50, scale=10, size=30)

# Productivity scores of the same employees after the training program
productivity_after = productivity_before + np.random.normal(loc=5, scale=3, size=30)

print(productivity_before.shape)

# Print the first 5 values for verification
print("Productivity Before Training:\n", productivity_before[:5])
print("\nProductivity After Training:\n", productivity_after[:5])
# Perform a paired t-test
t_statistic, p_value = stats.ttest_rel(productivity_after, productivity_before)

# Display results
print(f"\nT-Statistic: {t_statistic:.3f}")
print(f"P-Value: {p_value:.5f}")

alpha = 0.05  # Significance level
if p_value < alpha:
    print("Reject the null hypothesis")
else:
    print("Fail to reject the null hypothesis")
```

```python
# Dailychallenge:
# 1. Data import and cleaning
import pandas as pd

data_power_plant=pd.read_csv('global_power_plant_database.csv',sep=',')
data_power_plant.head()
data_power_plant.shape
data_power_plant.info()
print(data_power_plant.columns)


# Check for missing values
print("\nMissing values count:\n", data_power_plant.isna().sum())
```

```python
cols_to_fill = ['other_fuel1', 'other_fuel2', 'other_fuel3', 'wepp_id','generation_data_source','owner','url','source','geolocation_source']
cols_to_fill2=['commissioning_year','year_of_capacity_data','generation_gwh_2013','generation_gwh_2014','generation_gwh_2015','generation_gwh_2016','generation_gwh_2017','generation_gwh_2018','generation_gwh_2019','estimated_generation_gwh_2013','estimated_generation_gwh_2014','estimated_generation_gwh_2015','estimated_generation_gwh_2016','estimated_generation_gwh_2017']
data_power_plant[cols_to_fill] = data_power_plant[cols_to_fill].fillna("Unknow")

# # Fill missing values ​​in cols_to_fill2 with mean
data_power_plant[cols_to_fill2] = data_power_plant[cols_to_fill2].fillna(data_power_plant[cols_to_fill2].mean())

# # Check for missing values
print("\nMissing values count:\n", data_power_plant.isna().sum())
```

```python
print("\nSummary statistics for numerical columns:\n", data_power_plant.describe())
```

```python
# 2.Exploratory Data Analysis
# Group by country and count the number of power plants
country_distribution = data_power_plant.groupby('country')['name'].count().sort_values(ascending=False)

# Print the distribution of power plants by country
print("\nDistribution of power plants by country:\n", country_distribution)
```

```python
# Group by fuel type and count the number of power plants
fuel_distribution = data_power_plant.groupby('primary_fuel')['name'].count().sort_values(ascending=False)

# Print the distribution of power plants by fuel type
print("\nDistribution of power plants by fuel type:\n", fuel_distribution)
```

```python
# 3. Statistical analysis
stats = data_power_plant.groupby(['primary_fuel'])["capacity_mw"].agg(["mean", "std", "min", "max"])
print(stats)
```

```python
from scipy.stats import ttest_ind


# Extract power output values for Coal and Gas
coal_power = data_power_plant[data_power_plant["primary_fuel"] == "Coal"]["capacity_mw"].values
gas_power = data_power_plant[data_power_plant["primary_fuel"] == "Gas"]["capacity_mw"].values

#  Perform an independant  t-test and Print the results
t_stat, p_value =ttest_ind(coal_power, gas_power)

print(f"T-statistic: {t_stat:.2f}, P-value: {p_value:.4f}")
print(f"P-value: {p_value:.4f}")

# Interpret the p-value
alpha = 0.05
if p_value < alpha:
    print("\nThere is a significant difference in mean power output between fuel types.")
else:
    print("\nThere is no significant difference in mean power output between fuel types.")


```

```python
# 4.Time series analysis
data_power_plant['commissioning_year'] = data_power_plant['commissioning_year'].astype(int)

# Group data by 'commissioning_year' and fuel type to get the count
fuel_type_counts = data_power_plant.groupby(['commissioning_year', 'primary_fuel']).size().unstack(fill_value=0)

# Calculate the total number of plants per year
total_plants_per_year = fuel_type_counts.sum(axis=1)

# Normalize the fuel types counts to get the percentage distribution per year
fuel_type_percentage = fuel_type_counts.div(total_plants_per_year, axis=0) * 100

# Plot the evolution of fuel types over time
plt.figure(figsize=(12, 8))
fuel_type_percentage.plot(kind='area', stacked=True, figsize=(10,6), cmap='tab20')
plt.title("Evolution of Fuel Types for Power Generation Over Time")
plt.xlabel("Year")
plt.ylabel("Percentage of Total Power Plants")
plt.legend(title="Fuel Types", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()

# Calculate trends in fuel type distribution over time using NumPy
mean_fuel_type_distribution = fuel_type_percentage.mean(axis=0)
print("Mean Fuel Type Distribution Over Time:")
print(mean_fuel_type_distribution)
```

We can see a significant decrease in Hydro usage from 1900 to 2020 and an increase in Solar usage just little bit  before 2020

```python
#  6. Matrix Operations in Real-World Context:

matrix = data_power_plant[['capacity_mw', 'latitude', 'longitude']]

print(matrix[:5])
```

```python
# Transpose the Matrix
# Switch rows and columns:


matrix_T = matrix.T
print(matrix_T)
```

```python
# Shows relationships between attributes:

cov_matrix = np.cov(matrix_T)
print("Covariance Matrix:\n", cov_matrix)

# A high covariance between two attributes means they change together.
# If capacity and latitude have a high covariance, larger plants may be in certain regions.
```

```python
# Compute Eigenvalues & Eigenvectors
# Eigenvalues and eigenvectors help us reduce dimensions and find the strongest relationships:

eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)

print("Eigenvalues:\n", eigenvalues)
print("Eigenvectors:\n", eigenvectors)

# Eigenvalues tell us how much variance each principal component (axis of variation) captures.
# Eigenvectors define the direction of these variations.

# Dimensionality Reduction:
 # If one eigenvalue is much larger than others, it means most of the variation in the data is explained by that direction.
# Feature Importance:
#Helps determine which attribute (capacity, latitude, longitude) contributes the most to differences among power plants.
#Clustering & Classification:
# Useful for grouping power plants with similar characteristics
```
